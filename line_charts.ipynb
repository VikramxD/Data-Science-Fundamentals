```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
x=np.linspace(0,1000,30)
```

```
y=np.cos(x)+np.sin(x)
```

```
plt.plot(x,y)
plt.show()                                                                                               #always use if not using colab or notebook
```

```
plt.plot(x,y)
plt.xlabel("x input")
plt.ylabel("y output")
plt.title('First_Plot')
```
